```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from google.colab.patches import cv2_imshow
from collections import Counter
import itertools
```

```python
def pdf(a):
  leng=a.shape[0]*a.shape[1]
  ac=Counter(a.flatten()) #hist
  #pdf
  for x in ac:
    ac[x]=ac[x]/leng

  return ac

def cdf(ac):
  for x in sorted(ac):
    if x!=0:
      ac[x]=(ac[x]+ac[x-1])

  return ac

def eq(ac,a):
  b=[]
  for x in range(a.shape[0]):
    temp=[]
    for y in range(a.shape[1]):
      temp.append(ac[a[x][y]]*255)
    b.append(temp)

  return b
```

```python
def match(i,c2):
  for x in sorted(c2.values()):
    if x>i:return (list(c2.keys()))[list(c2.values()).index(x)]

  return 0

def matching(i1,i2):
  c1=cdf(pdf(i1))
  c2=cdf(pdf(i2))
  matched=[]
  for x in range(i1.shape[0]):
    temp=[]
    for y in range(i1.shape[1]):
      temp.append(match(c1[i1[x][y]],c2))
    matched.append(temp)

  return matched
```

```python
m=matching(i1,i2)
```
